# Noughts and Crosses (Tic-Tac-Toe)
A console game written in Python. Play against a friend or the computer.

**How to play:** Run all cells. Choose a mode, then take turns entering
a number 1–9 to place your mark. First to get three in a row wins.

### Global statistics and constants
The counters live outside all functions so they persist across games.
`WIN_COMBOS` stores the 8 winning lines as board index triples.

```python
import random  # Used by the computer player to pick random cells

# ----- Global statistics counters (exist outside all functions) -----
x_wins = 0        # Number of games won by player X
o_wins = 0        # Number of games won by player O
draws = 0         # Number of games ending in a draw
games_played = 0  # Total games completed

# The 8 possible winning lines, written as board index triples:
# 3 rows, 3 columns, 2 diagonals.
WIN_COMBOS = [
    (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
    (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
    (0, 4, 8), (2, 4, 6),             # Diagonals
]
```

### Board functions
Creating the board, displaying it, and a one-off cell number guide.

```python
def create_board():
    """Return a new board: a list of 9 spaces (empty cells)."""
    return [" "] * 9


def show_guide():
    """Print a one-off reference grid showing which number picks which cell."""
    print("Cells are numbered like this:")
    print(" 1 | 2 | 3 ")
    print("---+---+---")
    print(" 4 | 5 | 6 ")
    print("---+---+---")
    print(" 7 | 8 | 9 ")


def display_board(board):
    """Print the current board as a 3x3 grid."""
    print()
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print()
```

### Input handling
Both functions loop until the input is valid, so bad input can never
crash the game. `get_move` uses try/except to survive non-numeric input.

```python
def choose_mode():
    """Ask whether to play 2-player or vs the computer.

    Returns "2P" or "CPU".
    """
    while True:
        print("1 - Two players")
        print("2 - Play against the computer")
        choice = input("Choose a mode (1 or 2): ").strip()
        if choice == "1":
            return "2P"
        if choice == "2":
            return "CPU"
        print("Please enter 1 or 2.")


def get_move(board, player):
    """Ask the current player for a move and validate it.

    Keeps asking until the input is a number from 1-9 that
    points to a free cell. Returns the board index (0-8).
    """
    while True:  # Loop forever until we get a valid move
        choice = input(f"Player {player}, choose a cell (1-9): ")

        # Check 1: is it a number at all? int() crashes on "banana",
        # so we catch that crash instead of letting the game die.
        try:
            cell = int(choice)
        except ValueError:
            print("That's not a number - try again.")
            continue  # Skip back to the top of the loop

        # Check 2: is it within the board?
        if cell < 1 or cell > 9:
            print("Pick a number between 1 and 9.")
            continue

        # Check 3: is the cell still free?
        index = cell - 1  # Players type 1-9, but lists count from 0
        if board[index] in ("X", "O"):
            print("That cell is already taken - pick another.")
            continue

        return index  # Valid! Hand the index back to the caller
```

### Game logic
`check_win` loops through the 8 winning lines instead of using 8 if
statements. The computer tries to win, then tries to block, then
picks a random free cell. This mimics what a player would do.

```python
def check_win(board, player):
    """Return True if the given player has three in a row."""
    for a, b, c in WIN_COMBOS:
        if board[a] == player and board[b] == player and board[c] == player:
            return True
    return False  # Checked all 8 lines, no win


def check_draw(board):
    """Return True if every cell is taken (and nobody has won)."""
    for cell in board:
        if cell not in ("X", "O"):
            return False  # Found a free cell, so not a draw yet
    return True


def computer_move(board):
    """Choose a move for the computer (playing as O).

    Strategy: 1) win if possible, 2) block X's win, 3) random free cell.
    Returns the chosen board index.
    """
    free = [i for i in range(9) if board[i] == " "]

    # 1) If O can win right now, take that cell
    for i in free:
        board[i] = "O"                # Try the move...
        if check_win(board, "O"):
            board[i] = " "            # ...undo it and choose it for real
            return i
        board[i] = " "                # Undo and keep looking

    # 2) If X could win next turn, block that cell
    for i in free:
        board[i] = "X"
        if check_win(board, "X"):
            board[i] = " "
            return i
        board[i] = " "

    # 3) Otherwise pick any free cell at random
    return random.choice(free)
```

### Game loop and session
`play_game` runs one full game and updates the global stats.
`main` chooses the mode, then runs the replay loop.

```python
def play_game(mode):
    """Play one full game and update the global stats.

    mode is "2P" (two humans) or "CPU" (O is the computer).
    """
    global x_wins, o_wins, draws, games_played

    board = create_board()
    player = "X"  # X always goes first

    while True:  # The main turn loop
        display_board(board)

        # The computer plays O in CPU mode; humans type their moves
        if mode == "CPU" and player == "O":
            index = computer_move(board)
            print(f"Computer chooses cell {index + 1}")
        else:
            index = get_move(board, player)

        board[index] = player  # Place the mark

        # Check win BEFORE draw - a winning move can also fill the board
        if check_win(board, player):
            display_board(board)
            print(f"Player {player} wins!")
            if player == "X":
                x_wins += 1
            else:
                o_wins += 1
            break

        if check_draw(board):
            display_board(board)
            print("It's a draw!")
            draws += 1
            break

        # No win, no draw - other player's turn
        player = "O" if player == "X" else "X"

    games_played += 1


def show_stats():
    """Print the session statistics from the global counters."""
    print("\n----- Session stats -----")
    print(f"Games played: {games_played}")
    print(f"X wins:       {x_wins}")
    print(f"O wins:       {o_wins}")
    print(f"Draws:        {draws}")
    print("-------------------------")


def main():
    """Run the game session: welcome, mode choice, replay loop, final stats."""
    print("Welcome to Tic-Tac-Toe!")
    show_guide()
    mode = choose_mode()

    while True:  # The replay loop
        play_game(mode)
        show_stats()

        again = input("Play again? (y/n): ").strip().lower()
        if again != "y":
            print("Thanks for playing!")
            break
```

### Bonus Attempt

The next step would be to put real clickable buttons directly into the notebook. Ive had some experience doing this before so I'm going to attempt it.  The main deliverable for this project is above.

```python
import ipywidgets as widgets
from IPython.display import display

# ----- Interactive GUI version -----
# Reuses the same game logic; only inpu (clicks) and display change.

gui_board = create_board()
gui_state = {"player": "X", "over": False}

status = widgets.Label(value="Player X's turn")
stats_label = widgets.Label(value="")

mode_picker = widgets.ToggleButtons(
    options=[("Two players", "2P"), ("Vs computer", "CPU")],
    description="Mode:",
)

# Nine buttons = nine board cells
buttons = [
    widgets.Button(description=" ",
                   layout=widgets.Layout(width="64px", height="64px"))
    for _ in range(9)
]


def refresh_grid():
    """Copy the board list onto the button faces."""
    for i, btn in enumerate(buttons):
        btn.description = gui_board[i]


def update_stats_label():
    """Show the global counters under the board."""
    stats_label.value = (f"Games: {games_played} | X wins: {x_wins} | "
                         f"O wins: {o_wins} | Draws: {draws}")


def end_game(message, winner=None):
    """Mark the game over and update the global stats."""
    global x_wins, o_wins, draws, games_played
    gui_state["over"] = True
    status.value = message
    if winner == "X":
        x_wins += 1
    elif winner == "O":
        o_wins += 1
    else:
        draws += 1
    games_played += 1
    update_stats_label()


def make_move(index):
    """Place the current player's mark, then handle win/draw/switch."""
    player = gui_state["player"]
    gui_board[index] = player
    refresh_grid()

    if check_win(gui_board, player):
        end_game(f"Player {player} wins!", winner=player)
        return
    if check_draw(gui_board):
        end_game("It's a draw!")
        return

    gui_state["player"] = "O" if player == "X" else "X"
    status.value = f"Player {gui_state['player']}'s turn"


def handle_click(btn):
    """Respond to a human clicking a cell."""
    if gui_state["over"]:
        return  # Ignore clicks after the game ends
    index = buttons.index(btn)
    if gui_board[index] != " ":
        return  # Cell already taken - silently ignore

    make_move(index)

    # In CPU mode the computer replies immediately as O
    if (not gui_state["over"] and mode_picker.value == "CPU"
            and gui_state["player"] == "O"):
        make_move(computer_move(gui_board))


def new_game(_=None):
    """Reset the board and state for another game."""
    global gui_board
    gui_board = create_board()
    gui_state["player"] = "X"
    gui_state["over"] = False
    status.value = "Player X's turn"
    refresh_grid()


for btn in buttons:
    btn.on_click(handle_click)

new_game_btn = widgets.Button(description="New game", button_style="info")
new_game_btn.on_click(new_game)

grid = widgets.GridBox(
    buttons,
    layout=widgets.Layout(grid_template_columns="repeat(3, 68px)"),
)

display(widgets.VBox([mode_picker, grid, status, new_game_btn, stats_label]))
update_stats_label()
```
